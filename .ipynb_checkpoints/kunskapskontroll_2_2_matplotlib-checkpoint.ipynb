### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Plotting in Matplotlib can be done in either of two ways:

# 1. Object-Oriented Approach:
# This involves creating Figure and Axes objects explicitly and calling methods on them.
# It is the recommended approach because it provides more control and is better suited 
# for complex plots with multiple subplots or customizations.

import numpy as np
import matplotlib.pyplot as plt

# Example of the Object-Oriented Approach:
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sine Wave")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()

# 2. Pyplot State-Based Approach:
# This uses the state-based interface provided by matplotlib.pyplot.
# It is simpler and more concise but less flexible for complex plots.

# Example of the Pyplot State-Based Approach:
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# For most cases, especially when creating more advanced plots,
# the Object-Oriented Approach is recommended.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# ### Explanation of Figure, Axes, Axis, and Artist in Matplotlib:

# 1. Figure:
# - The **Figure** is the top-level container for all elements in a plot.
# - It acts as a blank canvas where one or more plots (axes) are drawn.
# - You can think of it as the entire window or page where the visualization is created.

# Example of creating a Figure:
import matplotlib.pyplot as plt
fig = plt.figure()  # Creates an empty figure
plt.show()

# 2. Axes:
# - The **Axes** is where the actual plot is created, such as a scatter plot, line plot, or bar plot.
# - It represents the plotting area within a Figure, including the data, labels, ticks, and gridlines.
# - A Figure can contain multiple Axes (subplots).

# Example of creating Axes within a Figure:
fig, ax = plt.subplots()  # Creates a figure with one set of axes
ax.plot([1, 2, 3], [4, 5, 6])  # Plots data on the axes
plt.show()

# 3. Axis:
# - An **Axis** is part of an Axes and refers to the x-axis or y-axis.
# - It includes the ticks, tick labels, and gridlines for one dimension of the plot.

# Example of customizing an Axis:
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_xlabel("X-axis")  # Customizing the x-axis
ax.set_ylabel("Y-axis")  # Customizing the y-axis
plt.show()

# 4. Artist:
# - An **Artist** is everything you see on the plot: lines, text, markers, etc.
# - Both Figures and Axes are Artists, as are elements like titles, labels, and legends.

# Example of an Artist:
fig, ax = plt.subplots()
line = ax.plot([1, 2, 3], [4, 5, 6], label="Line")  # The line is an Artist
ax.legend()  # The legend is another Artist
plt.show()

# In summary:
# - **Figure**: The overall container for the plot.
# - **Axes**: The specific area where data is plotted (subplots).
# - **Axis**: The individual x or y axis within an Axes.
# - **Artist**: Any visual element in the plot.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# ### Expected Input Data Types for Plotting in Matplotlib:
# 
# Matplotlib is flexible and can handle multiple data types for plotting. The most common ones are:
# 
# 1. **Lists or Tuples**:
# - Python native data structures like lists or tuples are commonly used.
# - Example:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]  # List
y = (10, 20, 30, 40, 50)  # Tuple
plt.plot(x, y)
plt.title("Using Lists and Tuples")
plt.show()

# 2. **NumPy Arrays**:
# - NumPy arrays are widely used for numerical computations and are ideal for plotting.
# - Example:
import numpy as np

x = np.linspace(0, 10, 100)  # NumPy array
y = np.sin(x)  # NumPy array
plt.plot(x, y)
plt.title("Using NumPy Arrays")
plt.show()

# 3. **Pandas Series**:
# - Matplotlib can directly plot Pandas Series for convenience.
# - Example:
import pandas as pd

data = pd.Series([5, 15, 25, 35, 45])
plt.plot(data)
plt.title("Using Pandas Series")
plt.show()

# 4. **Pandas DataFrame**:
# - When using DataFrames, you can specify columns to plot.
# - Example:
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [10, 20, 30, 40, 50]
})
plt.plot(data["x"], data["y"])
plt.title("Using Pandas DataFrame")
plt.show()

# ### Summary:
# - Common input types include Python lists, tuples, NumPy arrays, Pandas Series, and DataFrames.
# - NumPy arrays and Pandas DataFrames are recommended for efficiency and compatibility with numerical operations.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the range and function
x = np.linspace(-4, 4, 100)  # Create 100 points from -4 to 4
y = x**2  # Compute y = x^2

# Object-Oriented Approach
fig, ax = plt.subplots()  # Create a Figure and Axes
ax.plot(x, y, label="y = x^2")  # Plot the function
ax.set_title("Object-Oriented Approach: y = x^2")  # Set title
ax.set_xlabel("X-axis")  # Set x-axis label
ax.set_ylabel("Y-axis")  # Set y-axis label
ax.legend()  # Add a legend
plt.show()

# Pyplot Approach
plt.plot(x, y, label="y = x^2")  # Plot the function
plt.title("Pyplot Approach: y = x^2")  # Set title
plt.xlabel("X-axis")  # Set x-axis label
plt.ylabel("Y-axis")  # Set y-axis label
plt.legend()  # Add a legend
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Scatter plot (first subplot)
axes[0].scatter(random_data_x, random_data_y, alpha=0.5, label="Random Data")
axes[0].plot(x, y, color="red", label="y = x^2")
axes[0].set_title("Scatter Plot")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()

# Bar plot (second subplot)
axes[1].bar(names, values, color=['purple', 'red', 'orange', 'yellow', 'green'])
axes[1].set_title("Bar Plot")
axes[1].set_xlabel("Fruit")
axes[1].set_ylabel("Quantity")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
